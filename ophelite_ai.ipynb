```python
!pip install huggingface_hub -q

import os
from huggingface_hub import InferenceClient
from getpass import getpass

hf_token = getpass("HF Token: ")
client = InferenceClient(token=hf_token)
```

```python
def generate_ai_image(given_prompt):
  try:
      negative_prompt = "blurry, low quality, unrealistic, ugly"
      image = client.text_to_image(
          prompt=given_prompt,
          negative_prompt=negative_prompt,
          model="stabilityai/stable-diffusion-xl-base-1.0",
          height=512,
          width=768,
          num_inference_steps=90,
          guidance_scale=7.5,
          )

      return image

  except Exception as e:
      print(e)

my_image = generate_ai_image("A girl with curly, beeline honey hair and with glasses holding capsule coffee")
if my_image:
  my_image.save("my_image.png")
  display(my_image)
```
